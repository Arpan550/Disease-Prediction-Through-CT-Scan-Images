```python
import tensorflow as tf

model = tf.keras.models.load_model('/content/drive/MyDrive/Disease_Prediction/Cnn_ctscan_model.h5')
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()

with open('/content/drive/MyDrive/Disease_Prediction/model.tflite', 'wb') as f:
    f.write(tflite_model)
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()

with open('/content/drive/MyDrive/Disease_Prediction/quantized_model.tflite', 'wb') as f:
    f.write(tflite_model)
```
